```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore") 
```

```python
df = pd.read_csv('market-segmentasi.csv')

df
```

```python
df['MINIMUM_PAYMENTS'] = df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean())
df['CREDIT_LIMIT'] = df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean())
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop(columns=["CUST_ID"],axis=1,inplace=True)

df.head()
```

```python
plt.figure(figsize=(10,60))
for n in range(len(df.columns)):
    plt.subplot(17,1,n+1)
    sns.distplot(df[df.columns[n]], kde_kws={'color':'black','bw': 0.1,'lw':3,'label':'KDE'}, hist_kws={'color':'k'})
    plt.title(df.columns[n])
plt.tight_layout()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df= scaler.fit_transform(df)

scaled_df
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 3))
sns.histplot(df, ax=ax[0], kde=True, legend=False)
ax[0].set_title("Data Awal")
sns.histplot(scaled_df, ax=ax[1], kde=True, legend=False)
ax[1].set_title("Standarisasi Data")
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(scaled_df)

df_pca = pd.DataFrame(data=principal_components, columns=['PCA1', 'PCA2'])
print(df_pca)
```

# **K-MEANS**

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 15), wcss,)
plt.title('Elbow Method')
plt.xlabel('Values')
plt.ylabel('WCSS')
plt.show()
```

```python
from yellowbrick.cluster import KElbowVisualizer

model = KMeans(random_state=1)
visualizer = KElbowVisualizer(model, k=(1,15))

visualizer.fit(scaled_df)
visualizer.show()
plt.show()
```

```python
kmeans_model=KMeans(n_clusters=5)
kmeans_model.fit_predict(scaled_df)
df_pca_kmeans= pd.concat([df_pca,pd.DataFrame({'cluster':kmeans_model.labels_})],axis=1)
```

```python
print(df_pca_kmeans.columns)
```

```python
plt.figure(figsize=(8,8))
ax=sns.scatterplot(x = "PCA1", y = "PCA2", hue = "cluster" , data = df_pca_kmeans,palette = ['red','green','blue','black','yellow'])
plt.title("Clustering K-Means")
plt.show()
```

```python
cluster_centers = pd.DataFrame(data = model.cluster_centers_, columns = [df.columns])
cluster_centers
```

```python
cluster_centers = scaler.inverse_transform(cluster_centers)
cluster_centers = pd.DataFrame(data=cluster_centers,columns=[df.columns])
cluster_centers
```

```python
cluster_df = pd.concat([df,pd.DataFrame({'Cluster':kmeans_model.labels_})],axis=1)
cluster_df
```

```python
sns.countplot(x='Cluster', data=cluster_df)
```

```python
for c in cluster_df:
  grid = sns.FacetGrid(cluster_df, col='Cluster')
  grid.map(plt.hist, c)
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(scaled_df, kmeans_model.labels_)

print("Silhouette Score:", score)
```
